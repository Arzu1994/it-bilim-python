# **Третье домашнее задание**

Посредством OpenCV и Skimage сделать систему, которая будет искать ваше фамилию и имя в списке слушателей курса, подобно тому как мы распознавали цифры на 9й лекции.


На вход будет приниматься изображение произвольного размера, содержащее полный список слушателей курса. Если Ваша система нашла в списке Вас и только Вас, и обвела в рамочку - это 1 балл. Всего будет 20 изображений со списками.
А чтобы Ваша система Вас там не нашла - изображения мы будем слегка портить различными методами и их сочетаниями:
* Шум, блюр, различные цветовые гаммы
* Разного рода искажения
* Разные шрифты
* Грамматические ошибки
* Прочие секретные методики порчи списков посещения

**Ограничения на систему: **
* разрешается ипользовать только numpy, skimage и opencv, любые другие пакеты под запретом.
* система принимается в необученном виде + ровно 1 обучающее изображение 1000х1000 пикселей + 1 файл с разметкой.
* код для трейна и теста (вывода тестового изображения с рамкой) должен быть написан в jupyter notebook. 
* время обучения/теста не должно превышать 20 секунд

**За читерство и нарушения будем снимать баллы**

**Рекомендации:**
1. Не надо фокусироваться на сложных системах распознавания, оставьте это для курса машинного обучения.
